```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('starbucks_drinkMenu_expanded.csv')
df.head(20)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
print('Total Fat :', df[' Total Fat (g)'].unique())
```

```python
df.head()
```

```python
df['Beverage_category'].unique()
```

```python
df.calories = df[['Beverage_category', 'Calories']]
df.calories.columns = ['DrinkMenu', 'Calories']
df.calories
```

```python
plt.figure(figsize = (12,6), dpi =200)
sns.countplot( y = 'DrinkMenu', data = df.calories)
plt.show()
```

```python
df_beverage = pd.DataFrame(df['Beverage_category'].value_counts())
df_beverage['BeverageCategory'] = df_beverage.index
df_beverage.columns = ['NumberofDrinks', 'BeverageCategory']
df_beverage.reset_index(drop=True, inplace=True)
df_beverage
```

```python
df_beverage.head()
```

```python
df.calories.sort_values('Calories', ascending = False)
```

```python
plt.figure(figsize= (16,5), dpi=200)
plt.subplot(1,2,1)
sns.barplot(x='DrinkMenu',y='Calories', data = df.calories )
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Category Calories")

plt.subplot(1,2,2)
sns.boxplot(x = 'DrinkMenu', y='Calories', data = df.calories)
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Category Calories")
plt.show()
```

```python
plt.figure(figsize= (19,5), dpi=200)
plt.subplot(1,2,1)
sns.barplot(x='Beverage',y='Calories', data = df )
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Calories")

plt.subplot(1,2,2)
sns.boxplot(x = 'Beverage', y='Calories', data = df)
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Calories")
plt.show()
```
